```python
## import pyplot module from matplotlib package

import matplotlib.pyplot as plt
```

```python
## Plotting a simple line chart
## Let us take a bunch of X and Y values to plot
X = [ "CatA", "CatB", "CatC", "CatD", "CatE" ]
Y = [ 100, 50, 250, 340, 10]

plt.plot(X,Y)
plt.show()
```

```python
## Customizations to the line plot
X = [ "CatA", "CatB", "CatC", "CatD", "CatE" ]
Y = [ 100, 50, 250, 340, 10]

plt.xlabel('Categories')
plt.ylabel('Number of items in Categories')
plt.title('Categorical Counting Plot')
plt.plot(X,Y, marker='o', linestyle='--', color='g')
plt.grid(True)
plt.show()
```
